```python
from tkinter import *

def click(event):
    global scvalue
    text = event.widget.cget("text")
    print(text)
    if(text == "="):
        if(scvalue.get().isdigit()):
            value = int(scvalue.get())
        else:
            value = eval(screen.get())
        scvalue.set(value)
        screen.update()
    elif(text == "C"):
        scvalue.set("")
        screen.update()
    else:
        scvalue.set(scvalue.get() + text)
        screen.update()
    
root = Tk()
# root.geometry("100 x 200")  #width x height
root.title("Calculator")
root.minsize(350,300)
root.maxsize(350,300)

# user input block
scvalue = StringVar()
scvalue.set("")
screen = Entry(root, textvar = scvalue, font = "comicsansms 20 bold" )
screen.pack(fill=X, ipadx=8, padx=10, pady=10)

#9-0 numbers
f = Frame(root, bg="grey")
b = Button(f, text ="9", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, padx = 5, pady =5)
b.bind("<Button-1>", click)

b = Button(f, text ="8", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, pady =5)
b.bind("<Button-1>", click)

b = Button(f, text ="7", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, padx = 5, pady =5)
b.bind("<Button-1>", click)

b = Button(f, text ="+", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, padx = 5, pady =5)
b.bind("<Button-1>", click)
f.pack()

f = Frame(root, bg="grey")
b = Button(f, text ="6", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, padx = 5, pady =5)
b.bind("<Button-1>", click)

b = Button(f, text ="5", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, pady =5)
b.bind("<Button-1>", click)

b = Button(f, text ="4", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, padx = 5, pady =5)
b.bind("<Button-1>", click)

b = Button(f, text ="-", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, padx =5, pady =5)
b.bind("<Button-1>", click)
f.pack()

f = Frame(root, bg="grey")
b = Button(f, text ="3", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, padx = 5, pady =5)
b.bind("<Button-1>", click)

b = Button(f, text ="2", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, pady = 5)
b.bind("<Button-1>", click)

b = Button(f, text ="1", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, padx = 5, pady =5)
b.bind("<Button-1>", click)

b = Button(f, text ="*", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, padx =5, pady =5)
b.bind("<Button-1>", click)
f.pack()

f = Frame(root, bg="grey")
b = Button(f, text ="0", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, padx = 5, pady = 5)
b.bind("<Button-1>", click)

b = Button(f, text ="/", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, pady = 5)
b.bind("<Button-1>", click)

b = Button(f, text ="=", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, padx = 5, pady = 5)
b.bind("<Button-1>", click)

b = Button(f, text ="c", font = "comicsansms 15 bold", padx = 10, pady =5)
b.pack(side=LEFT, padx = 5, pady =5)
b.bind("<Button-1>", click)
f.pack()


root.mainloop()
```
