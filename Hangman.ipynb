```python
import random

hang = ["""

  +---+
  |   |
      |
      |
      |
      |
  =====""","""
  +---+
  |   |
  o   |
      |
      |
      |
  =====""","""
  +---+
  |   |
  o   |
  |   |
      |
      |
  =====""","""
  +---+
  |   |
  o   |
 /|   |
      |
      |
 =====""","""
  +---+
  |   |
  o   |
 /|\  |
      |
      |
 =====""","""
  +---+
  |   |
  o   |
 /|\  |
 /    |
      |
 =====""","""
  +---+
  |   |
  o   |
 /|\  |
 / \  |
      |
 ====="""
       ]

def generate_secret_word():
    words = ['rainbow', 'computer', 'science', 'programming','python', 'mathematics', 'player', 'condition','reverse', 'water', 
             'board', 'geeks']
    word = random.choice(words)
    play_game(word)

def play_game(word):
    dash = ["_"] *len(word)
    print(' '.join(dash))
    length = len(hang)
    count_wrong_guess = 0
    word_found = False
    while(count_wrong_guess < 7 and word_found == False ):    
        ch = input("Guess a letter ")
        if(ch in word):
            index_list = []
            for i in range(0, len(word)):
                if(word[i] == ch):
                    index_list.append(i)
            for i in index_list:
                dash[i] = ch   
        else:
            count_wrong_guess += 1
            print(hang[count_wrong_guess-1])
            
        if("_" not in dash):
            print("Yes you have guessed it right !!")
            word_found = True
            
        print(' '.join(dash))
        
    if(count_wrong_guess == 7):
        print("Oops! Your chances are over")
        print("The word is",word)
generate_secret_word()
    
```
